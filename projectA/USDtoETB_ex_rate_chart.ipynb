```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
dataSet = pd.read_csv('USD_ETB_Data_2010_2024.csv')
```

```python
dataSet.describe()
```

```python
dataSet.head()
```

```python
dataSet = dataSet.drop('Vol.', axis=1)
```

```python
dataSet.head()
```

```python
dataSet['Date'] = pd.to_datetime(dataSet['Date'])
```

```python

plt.plot(dataSet['Date'], dataSet['Price'], label='Price')
plt.plot(dataSet['Date'], dataSet['Open'], label='Open')
plt.xlabel('Date')
plt.ylabel('Exchange Rate')
plt.title('ETB_USD Currency Exchange Rates Over Time')
plt.legend()

plt.show()
```

```python

fig = px.line(dataSet, x='Date', y=['Price', 'Open'], title='ETB_USD Currency Exchange Rates')

fig.show()
```
