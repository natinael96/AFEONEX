```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df  = pd.read_csv('./health_facility_eth .csv')
```

```python
df.head()
```

```python
df['marker_size'] = 10
density_factor = 30
df['marker_size'] = df.groupby(['Lat', 'Long']).transform('count')['Country'] * density_factor
```

```python
fig = px.scatter_mapbox(
    df,
    lat="Lat",
    lon="Long",
    center={"lat": 9.0192, "lon": 40.2525},
    width=900,
    height=900,
    hover_data=["Admin1", 'Facility_n'],
    mapbox_style="open-street-map",

    zoom=5,
    opacity=0.75,
    color_continuous_scale="Viridis",
    size="marker_size",
)
#9.0192 N, 38.7525 E

fig.show()
```
